```python
import os

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
fpath = "../data/train/"
fname_list = os.listdir(fpath)

fname_list.sort()
fname_list
```

```python
fname = fname_list[0]

image = cv.imread(fpath + fname)

plt.figure(figsize=(5, 5))
plt.imshow(image)
plt.show()
```

```python
# Exmaple 1. Augmentation using Augraphy Default Pipeline.
from augraphy import *

pipeline = default_augraphy_pipeline()
image_augmented = pipeline(image)

# Display Image
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(image_augmented)
plt.show()
```

```python
# Example 2. Augmentation using User Defined Pipeline.
from augraphy import *

ink_phase = [DirtyDrum()]
paper_phase = [ColorPaper()]
post_phase = [BleedThrough()]

pipeline = AugraphyPipeline(
    ink_phase=ink_phase, paper_phase=paper_phase, post_phase=post_phase
)

image_augmented = pipeline(image)

# Display Image
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(image_augmented)
plt.show()
```

```python
# Example 3. Augmentation using Batches of Images (List).
from augraphy import *

# Make Pipeline for Augraphy
pipeline = default_augraphy_pipeline()

image_list = []
for i in range(100):
    image_list.append(cv.imread(fpath + fname_list[i]))

image_augmented_list = pipeline(image_list)

# Display Images
plt.figure(figsize=(25, 25))
image_rows = 1
image_columns = 10

for i in range(image_rows * image_columns):
    image_index = i + 1
    image_title = "Image_Agumented {}".format(image_index)
    plt.subplot(image_rows, image_columns, image_index)
    plt.title(image_title)
    # // plt.axis('off')
    plt.xticks([])  # x = None
    plt.yticks([])  # y = None
    plt.imshow(image_augmented_list[i])
plt.show()
```

```python
# Example 4. Augmentation using Archetype Pipelines (Predefined pipeline). There are 10 predefined pipelines now.
from augraphy import *
from augraphy.default.pipeline import *

pipeline = pipeline_archetype3()

image_list = []
for i in range(100):
    image_list.append(cv.imread(fpath + fname_list[i]))

image_augmented_list = pipeline(image_list)

# Display Images
plt.figure(figsize=(25, 25))
image_rows = 1
image_columns = 10

for i in range(image_rows * image_columns):
    image_index = i + 1
    image_title = "Image_Agumented {}".format(image_index)
    plt.subplot(image_rows, image_columns, image_index)
    plt.title(image_title)
    # // plt.axis('off')
    plt.xticks([])  # x = None
    plt.yticks([])  # y = None
    plt.imshow(image_augmented_list[i])
plt.show()
```
